```python
import pandas as pd
import os 
```

```python
df = pd.read_csv('resources/csv/dog_annotations_with_groups.csv')
df.head(10)
```

```python
img_folder_path = 'resources/stanford-dogs-dataset/images'
error_log = []
breed_name = ''

for _, row in df.iterrows():
    prev_breed = breed_name
    breed_name = row['breed_name']
    folder = row['folder']
    filename = row['filename']
    if filename == '%s.jpg':
        filename = filename.replace('%s',folder.split('-')[0])
        
    img_path = os.path.join(img_folder_path,folder,filename)
    
    if not os.path.exists(img_path):
        #print(f"⚠️ Missing file: {img_path}")
        if breed_name != prev_breed:
            print(breed_name)
        error_log.append(img_path)
        continue  # Skip missing images

print(len(error_log))
```
